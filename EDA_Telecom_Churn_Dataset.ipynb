#### Análise Exploratória dos Dados

Vamos começar checando a distribuição entre os gêneros.

```python
#Criando uma contagem
genero = df['gender'].value_counts()

#Cores para as barras
colors = ['steelblue', 'lightcoral']

#Criando subplot e definindo tamanho
fig, ax = plt.subplots(figsize=(7, 4))
#Criando gráfico de barras
ax.bar(genero.index, genero.values, color=colors)

#Rótulo eixo X
ax.set_xlabel('Gênero')
#Rótulo eixo Y
ax.set_ylabel('Contagem')

#Título
ax.set_title('Distribuição do gênero')

#Adicionando contagem em cada barra
for i, v in enumerate(genero.values):
    ax.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

Proporção dos gêneros

```python
labels = 'F', 'M'
sizes = [len(df[df['gender']=='F'])/len(df), len(df[df['gender']=='M'])/len(df)]
explode = (0, 0.1)
fig1, ax1 = plt.subplots(figsize=(7, 4))
ax1.pie(sizes, explode=explode, labels=labels, shadow=True, startangle=90, autopct='%1.1f%%')
ax1.axis('equal')
plt.title("Proporção para cada gênero", size = 16)
plt.show()
```

```python
#Calcular a media de salário estimado por gênero
median_salary_gender = df.groupby('gender')['estimated_salary'].median()

#cores
colors = ['steelblue', 'orange']
#Criando figura e eixo
fig, ax = plt.subplots(figsize=(8, 6))
#Plotando gráfico de barras
ax.bar(median_salary_gender.index, median_salary_gender.values, color = colors)

#Rótulos
ax.set_xlabel('Gênero')
ax.set_ylabel('Salário Estimado')
ax.set_title('Mediana de salário estimado VS Gênero')

#Textos dos rótulos
ax.text(0, median_salary_gender.values[0],
        f'{median_salary_gender.values[0]:.2f}', ha = 'center', va = 'bottom', fontsize = 12)

ax.text(1, median_salary_gender.values[1],
        f'{median_salary_gender.values[1]:.2f}', ha = 'center', va = 'bottom', fontsize = 12)

#Definindo limites dos eixos x e y
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(0, 160000)

#Calculando a diferença percentual
dif_percent = (median_salary_gender['F'] - median_salary_gender['M']) / median_salary_gender['M'] * 100

#Mensagem sobre diferença
ax.text(
    1.58, median_salary_gender.values[1],
    f'A mediana do salário estimado das mulheres é {dif_percent:.2f}% maior!',
    bbox = {'facecolor': 'lightsteelblue',
           'alpha': 1,
           'pad': 0.7,
           'edgecolor': 'none',
           'boxstyle': 'round'},
color = 'black',
fontsize = 11
)

plt.show()
```

#### Correlação entre variáveis

```python
sns.heatmap(df.corr());
```

```python
# Increase the size of the heatmap.
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
# Give a title to the heatmap. Pad defines the distance of the title from the top of the heatmap.
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
def display_correlation(df):
    r = df.corr(method="spearman")
    plt.figure(figsize=(10,6))
    heatmap = sns.heatmap(df.corr(), vmin=-1,
                      vmax=1, annot=True)
    plt.title("Spearman Correlation")
    return(r)
```

```python
r_simple=display_correlation(df)
```

```python
#separando as datas e as alocando em 3 novas colunas, ano, mês e dia de registro atribuindo a outra variavel
df2 = pd.concat([df.drop('date_of_registration', axis = 1),
          (df.date_of_registration.str.split('-').str[:3].apply(pd.Series)
          .rename(columns={0:'registration_year', 1:'registration_month', 2:'registration_day'}))], axis = 1)
```

```python
df2['registration_year'].value_counts()
```

```python
#Convertendo para tipo numérico
df2[['registration_year', 'registration_month', 'registration_day']] = df2[['registration_year', 'registration_month', 'registration_day']].apply(pd.to_numeric, errors='coerce')
```

```python
_ = df2.groupby('registration_year')['churn'].value_counts(sort=False).plot.bar()
```
